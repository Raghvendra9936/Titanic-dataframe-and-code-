# **Conditional Filter & Subset | DataFrame | Pandas | Python | Solved Exercise**

# **Import library**

```python
import pandas as pd
```

# **Import Titanic dataset and store as the pandas dataframe with name titanic**

```python
titanic = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Titanic.csv')
```

# **Create a new dataframe (female) from titanic of female passengers only and display first 8 rows**

```python
female = titanic.loc[titanic.sex == 'female']
```

```python
female.head(8)
```

# **Count number of female survived and not survived**

```python
female["survived"].value_counts()
```

```python
titanic['survived'].value_counts()
```

# **Count number of female survived and not survived as per passenger class**

```python
female[['survived','pclass']].value_counts()
```

```python
female.groupby(['survived','pclass']).count()
```

# **Select passengers with age > 18 years**

```python
titanic.loc[titanic[ 'age']>18, :]
```

# **Select passengers with age > 18 years and female**

```python
titanic[( titanic.sex == 'female' ) & (titanic[ 'age']>18 )]
```

# **Select passengers with age > 18 years and female and travelling by first class**

```python
titanic[( titanic.sex == 'female' ) & (titanic[ 'age']>18 ) & (titanic["pclass"] == 1)]
```

# **Select passengers with age > 18 years and female and travelling by first class and calculate their mean age**

```python
titanic[( titanic.sex == 'female' ) & (titanic[ 'age']>18 ) & (titanic["pclass"] == 1)  ]['age'].mean()
```

# **Select passengers with age > 18 years or female and travelling by first class**

```python
titanic[(titanic[ 'age']>18 ) | ( titanic.sex == 'female' )  & (titanic["pclass"] == 1)]
```

# **Select passengers with age > 18 years or female and travelling by first class and calculate their mean age**

```python
titanic[(titanic[ 'age']>18 ) | ( titanic.sex == 'female' )  & (titanic["pclass"] == 1)]['age'].mean()
```
